```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('measuring results/Sample_name-Water44.txt')
data = data.transpose()
```

```python
print(data.shape)
data_diff = np.diff(data[2,:])
data_diff = np.abs(data_diff)
print(data_diff.max())
hist, bin_edges = np.histogram(data_diff, bins=20)
pairs = np.stack((hist, bin_edges[1:]), axis=1)
```

```python
np.set_printoptions(suppress=True)
print(pairs)
#print(hist)
print(hist.sum())
```
